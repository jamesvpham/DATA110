```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# task 1

x = [264.1, 250.7, 1347.2, 318.7, 5.7, 38, 66.5]
y = ["Aggravated Assault", "Burglary", "Larceny", "Vehicle Theft", "Murder", "Rape", "Robbery"]

plt.bar(y, x, color = "maroon")
plt.grid(axis = "y", zorder = 0, linestyle = "-")
plt.ylabel("Offenses per 100,000 people (2023)");
plt.xlabel("Offense");
plt.title("Crime Rates (2023)")

# this bar graph is very cluttered, horizonal bar graph is preferred with these types of datasets
```

```python
# task 2

x = [-3.3, -8.1, -4.9, 12, -12, -9.8, -0.8]
y = ["Aggravated Assault", "Burglary", "Larceny", "Vehicle Theft", "Murder", "Rape", "Robbery"]

colors = ["crimson", "red", "maroon", "blue", "royalblue", "aqua", "green", "yellow"]

plt.barh(y, x, color = colors)
plt.grid(axis = "x", zorder = 0, linestyle = "-")
plt.xlabel("Change in Crime Rates (2023)");
plt.ylabel("Offense");
plt.title("Crime Rates (2023)")

'''
horizonal graphs are better suited for these types of
data sets since the names will make the graph look
cluttered if the names of each crime are on the x axis.

change in crime rates is more accurate and paints
the full picture, whereas showing the absolute numbers
may have people thinking that this place is crime ridden.

the change in percentage of crimes shows the full context,
most crimes are being dialed down and less than previous years.
'''
```

```python
# task 2 (continued)

x = [-3.3, -8.1, -4.9, 12, -12, -9.8, -0.8]
y = ["Aggravated Assault", "Burglary", "Larceny", "Vehicle Theft", "Murder", "Rape", "Robbery"]

plt.barh(y, x, color = "darkblue")
plt.grid(axis = "x", zorder = 0, linestyle = "-")

plt.xlabel("Percentage of change in Crime Rates (2023)");
plt.ylabel("Offense");
plt.title("Crime Rates (2023)")

'''
i feel like it helps with readability when all the colors are the same.
sometimes it's better to keep these graphs simple with one color to ease
on the eyes.
'''
```

```python
# task 2 (final)

x = [-3.3, -8.1, -4.9, 12, -12, -9.8, -0.8]
y = ["Aggravated Assault", "Burglary", "Larceny", "Vehicle Theft", "Murder", "Rape", "Robbery"]
colors = ["darkblue", "darkblue", "darkblue", "maroon", "darkblue", "darkblue", "darkblue"]

plt.barh(y, x, color = colors)
plt.grid(axis = "x", zorder = 0, linestyle = "--")

plt.xlabel("Percentage of change in Crime Rates (2023)");
plt.ylabel("Offense");
plt.title("Crime Rates (2023)")

'''
for the final adjustment, i changed the gridlines (doesn't make much of a difference)
and i also changed one of the colors. the reason why i changed vehicle
theft's color is to show viewers the contrasting statistic.

vehicle theft is the only statistic that has increased, and changing its
color highlights it in this graph.
'''
```

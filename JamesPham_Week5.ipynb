```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# View the first few rows of the dataset
titanic.head()
```

This titanic dataset shows rows of people and has multiple columns of information about these individuals. There are columns that show the person's sex, age, class, whether they are alive, whether they are alone, and more.

```python
# part 2 task 2

plt.hist(titanic['fare'], bins = 15)
plt.xlabel("Fare")
plt.ylabel("amount")
plt.show()
```

```python
# part 2 task 2

sns.kdeplot(data = titanic["fare"], fill = True)
plt.xlim(0, 560)
plt.show()
```

The majority of people on the titanic paid between 0-100$ in fares. The outliers have fares that go out to the 500s.

```python
# part 2 task 3

fare_alive = titanic[titanic["alive"] == "yes"]["fare"]
fare_dead = titanic[titanic["alive"] == "no"]["fare"]

sns.kdeplot(fare_alive, fill = True, color = "blue", label = "alive")
sns.kdeplot(fare_dead, fill = True, color = "red", label = "dead")
plt.xlim(0, 600)
plt.legend()
plt.show()
```

The graph suggests that there is a significant increase in density between the amount of people who died and the amount of people who survived that paid fares between 0-100$. Additionally, those who paid higher fees were more likely to survive. This shows that those that were of a lower socioeconomic status were more likely to die than those of higher socioeconomic status.

```python
# part 3 kde

fare_cl1 = titanic[titanic["pclass"] == 1]["fare"]
fare_cl2 = titanic[titanic["pclass"] == 2]["fare"]
fare_cl3 = titanic[titanic["pclass"] == 3]["fare"]

sns.kdeplot(fare_cl1, fill = True, color = "blue", label = "class 1")
sns.kdeplot(fare_cl2, fill = True, color = "red", label = "class 2")
sns.kdeplot(fare_cl3, fill = True, color = "green", label = "class 3")

plt.xlim(0, 600)
plt.legend()
plt.show()
```

```python
# part 3 histogram

plt.figure(figsize = (10, 6))
plt.hist([fare_cl1, fare_cl2, fare_cl3], bins = 20, color = ["blue", "red", "green"], label = ["class 1", "class 2", "class 3"])
plt.xlabel("fare")
plt.ylabel("people/density")
plt.legend()

plt.show()
```

From these graphs, it is evident that first class passengers are observedly the only few that paid higher fares. Moreover, there are significantly more third class passengers that paid 0-100 than first and second class passengers.

```python
# part 4 task 5

age_alive = titanic[titanic["alive"] == "yes"]["age"]
age_dead = titanic[titanic["alive"] == "no"]["age"]

sns.kdeplot(age_alive, fill = True, color = "blue", label = "alive")
sns.kdeplot(age_dead, fill = True, color = "red", label = "dead")

plt.legend()
plt.xlim(0, 100)
plt.show()
```

From this graph, it seems like the age of people that died are mostly between 20-40s. It is the same way with the age of people that survived. It also seems like there are more people that died than survived in the 60-80s range. This could mean that the oldest people were more likely to have not survived.

```python
# part 4 task 6

male_alive = titanic[(titanic["sex"] == "male") & (titanic["alive"] == "yes")]["age"]
male_died = titanic[(titanic["sex"] == "male") & (titanic["alive"] == "no")]["age"]


plt.figure(figsize = (12, 6))

plt.hist([male_alive], bins = 20, label = ["alive"], alpha = 0.5, color = ["blue"], orientation = 'vertical', weights = [-1] * len(male_alive))
plt.hist([male_died], bins = 30, label = ["dead"], alpha = 0.5, color = "red", orientation = 'vertical')

plt.yticks([-60, -40, -20, 0, 20, 40, 60], [60, 40, 20, 0, 20, 40, 60])
plt.ylabel("Amount")
plt.xlabel("Age")
plt.legend()

plt.show()
```

```python
# part 4 task 6

female_alive = titanic[(titanic["sex"] == "female") & (titanic["alive"] == "yes")]["age"]
female_died = titanic[(titanic["sex"] == "female") & (titanic["alive"] == "no")]["age"]


plt.figure(figsize = (12, 6))

plt.hist([female_alive], bins = 20, label = ["alive"], alpha = 0.5, color = ["blue"], orientation = 'vertical', weights = [-1] * len(female_alive))
plt.hist([female_died], bins = 30, label = ["dead"], alpha = 0.5, color = "red", orientation = 'vertical')

plt.yticks([-60, -40, -20, 0, 20, 40, 60], [60, 40, 20, 0, 20, 40, 60])
plt.ylabel("Amount")
plt.xlabel("Age")
plt.legend()

plt.show()
```

From these graphs, it seems like there are more male passengers that did not survive compared to female passengers. The most that did not survive were male passengers that were between the ages 20-40.

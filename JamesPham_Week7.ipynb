```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

iris = sns.load_dataset('iris')
iris.head()
```

```python
# pairplot of species of all flowers
sns.pairplot(iris, hue = "species", diag_kind = "kde")
plt.title("Pairplot of Species of Flowers") # i don't know why the title is in the middle of the graphs

plt.show()
```

This pair plot shows the many varieties of graphs between the datatypes of petals and the sepal. With the help of the color coded legend, we can see that the setosa flower has the smallest petal width of the bunch. The next one in petal width size is the versicolor species. Lastly, the largest petal width is the virginica species which is noticeably the biggest one being in the high Y axis of each graph in petal width.

We can conclude that the smallest petal species is the setosa species because the 3rd graph in the last row shows that the setosa species are mostly within the bottom left corner of the graph.

```python
# bar plot of sepal length by species
sepall_mean = iris.groupby('species')['sepal_length'].mean()
plt.barh(sepall_mean.index, sepall_mean.values)
plt.title("Average Sepal Length by Species")
plt.xlabel("Average Sepal Length by Inch")
plt.grid(axis = "x") # grid for the x axis

plt.show()
```

This bar graph shows the average sepal length in mean. We can see that the virginica species has the largest average in sepal length.

```python
# box plot of petal width distribution by species
sns.boxplot(x = iris["species"], y = iris["petal_width"])
plt.title("Petal Width Distribution by Species")
plt.grid(axis = "y") # grid for the y axis
plt.ylabel("Petal Width by Inches")

plt.show()
```

Once again, the virginica seems to be the largest species here. The virginica species has the largest petal width since most of its flowers are between 1.8-2.3. inches. With the highest of its flowers being as big as 2.5 inches.

The setosa species is in the very low range, with a few outliers of it being higher than 0.5 inches.

```python
# bar plot of average petal length
petall_mean = iris.groupby('species')['petal_length'].mean()

plt.barh(petall_mean.index, petall_mean.values)
plt.title("Average Petal Length of Species")
plt.xlabel("Petal Length by Inches")
plt.grid(axis = "x")
plt.show()
```

This bar graph shows how small the setosa species average petal length is. Once again, the virginica species is the biggest one here.

```python
# violin plot of sepal length distribution by species
sns.violinplot(x = iris["species"], y = iris["sepal_length"])
plt.title("Sepal Length Distribution by Species")
plt.grid(axis = "y")
plt.ylabel("Sepal Length by Inches")

plt.show()
```

This graph shows a very interesting pattern. It's almost like a smooth slope can be created showing the difference in length between the species. I assume that the setosa species has a high volume of flowers within that 5 inch length area due to the width of the violin. Another thing that I observed is that the tip of virginica stretches high and low. This highlights the outliers within the virginica species. A flower or two might be shorter in sepal length compared to the others, and there's obviously a few that are much longer in length past the 8 inch mark.

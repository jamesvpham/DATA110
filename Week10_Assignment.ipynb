```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Reben80/Data110-22016/refs/heads/main/dataset/Daily_Arrests_20241001%20(1).csv')
theft_df = df[df["OFFENSE"].str.contains("Theft", case = False, na = False)]
theft_df.head()
```

```python
theft_df.describe()
```

```python
# age distribution of theft counts
plt.hist(theft_df["AGE"], bins = 15)
plt.xlabel("Age")
plt.ylabel("Amount")
plt.xlim(5,70)
plt.show()
```

There are a total of 96 offenses of theft. The age of theft offenses range from about 18 to 68. The most thefts recorded by age range are between 20-30.

```python
theft_df["ARREST DATE"] = pd.to_datetime(df["ARREST DATE"])
theft_df["dayOfWeek"] = theft_df["ARREST DATE"].dt.dayofweek

day_counts = theft_df["dayOfWeek"].value_counts()
day_names = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

plt.barh(day_names, day_counts)
plt.xlabel("Counts of Theft")
plt.show()
```

Majority of the thefts seem to be conducted on the weekdays, but the day with the most thefts is Monday. I don't know why it would be a Monday, but I feel like there's not a lot of theft during the weekend because some stores could be closed.

The only theory I could propose is that these thefts are being committed by people without jobs, so they are more likely to be busy on weekends with plans than on a weekday like Monday.

```python
# getting the counts of theft using a horizontal bar graph
theft_counts = theft_df["OFFENSE"].value_counts()
plt.barh(theft_counts.index, theft_counts)
plt.show()
```

The most common type of theft, by a significant margin, is theft that causes between 100-1,500$ in damages or lost goods.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import folium
from folium.plugins import HeatMap
```

```python
df = pd.read_csv("/content/BPD_Arrests.csv")
df.head()
```

```python
# get center of map

map_center = [39.2904, -76.6122] # center of baltimore
m = folium.Map(location = map_center, zoom_start = 11)
```

```python
# create male and female dfs

male_df = df[df["Gender"].str.contains("M", na = False)]
female_df = df[df["Gender"].str.contains("F", na = False)]

male_arrests = male_df.dropna(subset = ["Latitude", "Longitude"])
female_arrests = female_df.dropna(subset = ["Latitude", "Longitude"])

female_arrests.head()
```

```python
# get coordinates

male_values = male_arrests[["Latitude", "Longitude"]].values
female_values = female_arrests[["Latitude", "Longitude"]].values
```

```python
# create layers for each

male_layer = folium.FeatureGroup(name="Male Arrests Heatmap")
female_layer = folium.FeatureGroup(name="Female Arrests Heatmap")
```

```python
# add to heatmap and add to combined map

HeatMap(male_values).add_to(male_layer)
HeatMap(female_values).add_to(female_layer)

male_layer.add_to(m)
female_layer.add_to(m)

# add layer control to map
folium.LayerControl().add_to(m)
```

```python
# create file path and export file

map_path = "male_female_heatmap.html"
m.save(map_path)
map_path
```

At a glance, it seems like there are more male arrests than female arrests.

```python
# libraries

import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

```python
# basic line plot

x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1
plt.plot(x, y)
```

```python
# multiple lines

x = np.linspace(-2, 4, 100)
plt.plot(x, x**3, x, np.exp(x), x, np.log(x))
plt.legend(['f(x)', 'g(x)', 'h(x)'])
```

```python
# customized styles

x = np.linspace(-10, 10, 100)

plt.plot(x, x**2, color = "crimson", linestyle = "--")
plt.plot(x, x**3, color = "royalblue")
plt.plot(x, x**4, color = "green", linestyle = "-.")
plt.plot(x, x**5, color = "yellow", linestyle = ":")

plt.legend(['f(x)', 'g(x)', 'h(x)', 'i(x)'])
# print(plt.style.available)
```

```python
# axes limits and scaling

x = np.linspace(-10, 10, 100)
plt.plot(x, 1/x**2+1)
plt.xlim(-5, 5)
plt.ylim(-1, 10)
plt.axis('tight')
```

```python
# axes limits and scaling 2

x = np.linspace(-10, 10, 100)
plt.plot(1/x**2+1, x)
plt.xlim(-10, 20)
plt.ylim(-5, 5)
plt.axis('equal')
```

```python
# labels and legends

x = np.linspace(-10, 10, 100)
plt.plot(x, np.sqrt(x), label = "f(x)")
plt.plot(x, x**(1/3), label = "g(x)")

# titles and labels for the graph
plt.title("Graph of f(x) and g(x)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```

```python
# basic scatter plot

x = np.random.rand(10)
y = np.random.rand(10)

# plot 1
plt.plot(x, y, marker = 'o')
# plot 2 (must remove plot 1's line in order to see plot 2 which is just dots and no lines)
plt.scatter(x, y)
```

```python
# customizing scatter plot

x = np.linspace(0, 4*np.pi)
plt.plot(x, np.sin(x), marker = ".", ms = 10, mec = "red", mfc = "black", mew = 2, linewidth = 2)
```

```python
# combined plot

x = np.linspace(-np.pi, np.pi)
plt.plot(x, np.cos(x), marker = 'o', mec = "red", color = "green")
```

```python
# custom plot
plt.style.use('classic')

x = np.linspace(-np.pi, np.pi)
plt.plot(x, np.tan(x), label = "f(x)", marker = "o", mec = "aqua")
plt.plot(x, np.arctan(x), linestyle = "--", color = "crimson", label = "i(x)")

# graph titles/labels

plt.xlabel("x")
plt.ylabel("y")
plt.legend()

'''
there is no particular reason why i chose these different functions,
the style that i incorporated makes the plot look old-school and very
outdated.
'''
```
